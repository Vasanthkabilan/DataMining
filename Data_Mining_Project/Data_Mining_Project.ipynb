```python
def visualize_correlation_heatmap(data_df, plot_width, size=None):
    # Drop columns with NaN
    data_df = data_df.dropna(axis='columns', how='any')
    
    # Keep columns where there are more than 1 unique values
    data_df = data_df.loc[:, data_df.nunique() > 1]
    
    if data_df.shape[1] < 2:
        print(f'No correlation plots shown: The number of non-NaN or constant columns ({data_df.shape[1]}) is less than 2')
        return
    
    if size:
        data_df = data_df.sample(n=size, random_state=42)
    
    # Calculate correlation matrix
    correlation_matrix = data_df.corr()
    
    # Ploting correlation matrix using Seaborn heatmap
    plt.figure(figsize=(plot_width, plot_width))
    sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f", linewidths=.5)
    
    plt.title('Correlation Matrix Heatmap', fontsize=15)
    plt.show()


```

```python
# Visualizing heatmap with size of 1000
visualize_correlation_heatmap(train_data, plot_width=10, size=1000)
visualize_correlation_heatmap(test_data, plot_width=10, size=1000)
```

```python
def visualize_scatter_density(data_df, plot_size, text_size):
    data_df = data_df.select_dtypes(include=[np.number])  # Keep only numerical columns
    # Remove rows and columns that would lead to df being singular
    data_df = data_df.dropna(axis='columns')
    data_df = data_df.loc[:, data_df.nunique() > 1]  # Keep columns where there are more than 1 unique values
    
    column_names = list(data_df)
    if len(column_names) > 10:  # Reduce the number of columns for matrix inversion of kernel density plots
        column_names = column_names[:10]
    
    data_df = data_df[column_names]
    
    scatter_matrix = pd.plotting.scatter_matrix(data_df, alpha=0.75, figsize=[plot_size, plot_size], diagonal='kde')
    correlations = data_df.corr().values
    
    for i, j in zip(*plt.np.triu_indices_from(scatter_matrix, k=1)):
        scatter_matrix[i, j].annotate('Corr. coef = %.3f' % correlations[i, j], (0.8, 0.2), xycoords='axes fraction', ha='center', va='center', size=text_size)
    
    plt.suptitle('Scatter and Density Plot')
    plt.show()
```

```python
visualize_scatter_density(train_data, plot_size=10, text_size=10)
visualize_scatter_density(test_data, plot_size=10, text_size=10)
```

```python
# importing sklearn library
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
# Importing 'SimpleImputer'
from sklearn.impute import SimpleImputer
# Importing SelectKBest
from sklearn.feature_selection import SelectKBest, f_classif
```

```python
train_data =pd.read_csv('mitbih_train.csv', header=None)
test_data =pd.read_csv('mitbih_test.csv', header=None)
```

```python
# Assuming the target Variable
X_train = train_data.iloc[:, :-1]
y_train = train_data.iloc[:, -1]

X_test = test_data.iloc[:, :-1]
y_test = test_data.iloc[:, -1]
```

```python
# Handling Missing Values
imputer = SimpleImputer(strategy='mean')
X_train = imputer.fit_transform(X_train)
X_test = imputer.transform(X_test)
```

```python
# Standardize features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Feature selection
k_best = 10
selector = SelectKBest(f_classif, k=k_best)
X_train_selected = selector.fit_transform(X_train, y_train)
X_test_selected = selector.transform(X_test)
```

```python
# Printing the selected feature
selected_feature_indices = selector.get_support(indices=True)
print(f'Selected feature indices: {selected_feature_indices}')
```

```python
# Train a model (for demonstration purposes, using a RandomForestClassifier)
model = RandomForestClassifier()
model.fit(X_train_selected, y_train)
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test_selected)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Model Accuracy: {accuracy}')
```
